```python

# ========================================================
# COMPARE: ️ DDL vs DML vs DQL — Key Differences
# SESSION: Session 1 — SQL Foundations, DDL, DML & Constraints
# ========================================================

'''
[DDL — Data Definition Language]
  - Deals with database structure/schema
  - Commands: CREATE, ALTER, DROP, TRUNCATE
  - Changes are auto-committed — cannot be rolled back in MySQL
  - Affects the structure of tables/databases, not data directly
[DML — Data Manipulation Language]
  - Deals with data inside tables
  - Commands: INSERT, UPDATE, DELETE
  - Changes can be rolled back within a transaction
  - Affects the actual rows of data
[DQL — Data Query Language]
  - Deals with retrieving/reading data
  - Command: SELECT
  - Does NOT modify anything — read only
  - Returns data as a result-set
[Key Differences]
  - DDL operates on database structure; DML/DQL operate on data
  - DDL changes cannot be rolled back; DML changes can be rolled back
  - DQL only reads — no modification; DDL/DML modify structure or data
  - TRUNCATE is DDL (not rollback-able in MySQL); DELETE is DML (rollback-able)
  NOTE: ⚠️ Common Exam Trap: Students often classify SELECT as DML. Technically SELECT is DQL. However, in some syllabi it is grouped under DML — check your syllabus.
'''


# ========================================================
# COMPARE: ️ DROP vs TRUNCATE vs DELETE — Key Differences
# SESSION: Session 1 — SQL Foundations, DDL, DML & Constraints
# ========================================================

'''
[DROP TABLE]
  - Permanently removes the entire table — structure AND data
  - DDL command — auto-committed, cannot be rolled back
  - Table no longer exists after DROP
[TRUNCATE TABLE]
  - Removes all rows inside a table but keeps the table structure
  - DDL command — in MySQL, cannot be rolled back
  - Faster than DELETE because it does not log individual row deletions
  - Resets AUTO_INCREMENT counters
[DELETE FROM]
  - Removes specific rows based on WHERE condition (or all rows if no WHERE)
  - DML command — can be rolled back within a transaction
  - Slower than TRUNCATE because each row deletion is logged
  - Does NOT reset AUTO_INCREMENT counters
[Key Differences]
  - DROP removes the table itself; TRUNCATE removes data but keeps structure; DELETE removes specific rows
  - DROP and TRUNCATE are DDL (not rollback-able in MySQL); DELETE is DML (rollback-able)
  - DELETE uses WHERE clause; TRUNCATE and DROP do not
  - TRUNCATE is the fastest; DELETE is the most flexible (selective deletion)
  NOTE: ⚠️ Common Mistake: Thinking TRUNCATE can be rolled back in MySQL — it cannot. Only DELETE inside a transaction can be rolled back.
'''


# ========================================================
# COMPARE: ️ CHAR vs VARCHAR — Key Differences
# SESSION: Session 1 — SQL Foundations, DDL, DML & Constraints
# ========================================================

'''
[CHAR (Fixed Length)]
  - Fixed-length storage — always uses the specified number of characters
  - If data is shorter than defined length, MySQL pads with spaces
  - Maximum length: 8,000 characters
  - Faster for fixed-size data (e.g., country codes, gender codes)
  - Wastes storage if data varies in length
[VARCHAR (Variable Length)]
  - Variable-length storage — stores only the actual number of characters used
  - No padding — efficient for strings of varying length
  - Maximum length: 8,000 characters
  - Slower than CHAR for fixed-size data due to overhead
  - Better storage efficiency for names, addresses, descriptions
[Key Differences]
  - CHAR: fixed storage allocation; VARCHAR: dynamic storage allocation
  - CHAR pads with spaces; VARCHAR does not
  - Use CHAR when values always have the same length (e.g., state code 'MH', 'KA')
  - Use VARCHAR when values have varying lengths (e.g., names, cities)
  NOTE: ⚠️ Common Exam Trap: Both CHAR and VARCHAR have the same maximum limit (8,000 chars). Students often confuse them as having different limits — the limit is the same; only the storage behaviour differs.
'''


# ========================================================
# COMPARE: ️ PRIMARY KEY vs UNIQUE KEY — Key Differences
# SESSION: Session 1 — SQL Foundations, DDL, DML & Constraints
# ========================================================

'''
[PRIMARY KEY]
  - Uniquely identifies each record in a table
  - Cannot contain NULL values
  - Only ONE PRIMARY KEY per table
  - Automatically creates a clustered index
  - Is both NOT NULL and UNIQUE
[UNIQUE KEY]
  - Ensures all values in the column are different
  - CAN contain NULL values (only one NULL allowed in MySQL)
  - A table can have MULTIPLE UNIQUE keys
  - Creates a non-clustered index by default
[Key Differences]
  - NULL: Primary Key rejects NULL; Unique Key allows NULL
  - Count: Only 1 Primary Key per table; multiple Unique Keys allowed
  - Purpose: Primary Key = main identifier; Unique Key = prevent duplicates in other columns
  - Primary Key automatically implies Unique — but Unique does not imply Primary
  NOTE: ⚠️ Common Exam Trap: Students say Unique Key cannot have NULL — in MySQL, one NULL is allowed in a UNIQUE column.
'''


# ========================================================
# COMPARE: ️ CANDIDATE KEY vs PRIMARY KEY — Key Differences
# SESSION: Session 1 — SQL Foundations, DDL, DML & Constraints
# ========================================================

'''
[Candidate Key]
  - Any column or combination of columns that CAN uniquely identify a record
  - Multiple candidate keys can exist in one table
  - Each candidate key qualifies to be a primary key
  - Example: In a student table, both StudentID and Email could be candidate keys
[Primary Key]
  - The ONE candidate key that is SELECTED to be the main identifier
  - Only one primary key per table
  - Cannot contain NULL; must be unique
  - All candidate keys that are NOT chosen become Alternate Keys
[Key Differences]
  - Every Primary Key is a Candidate Key; NOT every Candidate Key is a Primary Key
  - There can be multiple Candidate Keys; only one Primary Key
  - Unchosen candidate keys become Alternate Keys
'''


# ========================================================
# COMPARE: ️ CHANGE vs MODIFY Clause in ALTER — Key Differences
# SESSION: Session 1 — SQL Foundations, DDL, DML & Constraints
# ========================================================

'''
[CHANGE Clause]
  - Can rename a column AND change its data type/constraints
  - Requires: old column name, new column name, new data type
  - Syntax: ALTER TABLE t CHANGE old_col new_col datatype;
  - More powerful — superset of MODIFY
[MODIFY Clause]
  - Can only change data type or constraints — CANNOT rename the column
  - Requires: current column name, new data type
  - Syntax: ALTER TABLE t MODIFY col_name datatype;
  - Simpler — use when only definition needs to change
[Key Differences]
  - CHANGE can rename; MODIFY cannot rename
  - CHANGE requires both old and new column name; MODIFY only needs current name
  - MODIFY does everything CHANGE can, but without renaming the column
  - Use CHANGE when renaming; use MODIFY when changing type/constraints only
  NOTE: ⚠️ Common Mistake: Using MODIFY to rename a column — it will fail. Always use CHANGE for renaming.
'''


# ========================================================
# COMPARE: ️ Logical Schema vs Physical Schema — Key Differences
# SESSION: Session 1 — SQL Foundations, DDL, DML & Constraints
# ========================================================

'''
[Logical Schema]
  - Designed based on business requirements
  - Represents the conceptual/logical design of the database
  - May not have column data types defined
  - Used for business analysis and planning
  - Does not describe how data is physically stored
[Physical Schema]
  - Represents the actual blueprint of the relational database
  - Includes data types, primary keys, foreign keys, and constraints
  - Describes how data is stored in actual physical storage
  - Designed by database administrators/designers
[Key Differences]
  - Logical = what; Physical = how
  - Logical is for planning; Physical is for implementation
  - Physical schema has specific data types; Logical may not
'''


# ### Session 2 — WHERE Predicates, Set Ops & Functions


# ========================================================
# COMPARE: ️ UNION vs UNION ALL — Key Differences
# SESSION: Session 2 — WHERE Predicates, Set Ops & Functions
# ========================================================

'''
[UNION]
  - Combines results of two SELECT statements
  - Removes duplicate rows — each unique row appears only once
  - Slower — requires an extra deduplication step (sort + compare)
  - Use when you need a distinct/unique combined result set
[UNION ALL]
  - Combines results of two SELECT statements
  - Keeps all rows including duplicates
  - Faster — no deduplication processing needed
  - Use when duplicates are acceptable or expected, or when performance matters
[Key Differences]
  - Duplicate rows: UNION removes them; UNION ALL keeps them
  - Performance: UNION ALL is faster because no deduplication step
  - Result size: UNION result ≤ UNION ALL result (for the same queries)
  - Use UNION when results must be distinct; use UNION ALL when you want all rows or when you know there are no duplicates
  NOTE: ⚠️ Common Exam Trap: Students assume UNION ALL always gives more rows than UNION. This is only true when duplicates exist. If there are no duplicates, both give the same number of rows — but UNION ALL is still faster.
'''


# ========================================================
# COMPARE: ️ % Wildcard vs _ Wildcard — Key Differences
# SESSION: Session 2 — WHERE Predicates, Set Ops & Functions
# ========================================================

'''
[% (Percent) Wildcard]
  - Matches zero or more characters of any type
  - More flexible — can match a word, a sentence, or nothing at all
  - Examples: 'A%' matches A, Abc, Abcdef; '%A%' matches anything containing A
[_ (Underscore) Wildcard]
  - Matches exactly one character
  - More precise — use when you know the exact length of the unknown portion
  - Multiple underscores can be chained: '__' = exactly 2 chars, '___' = exactly 3
  - Examples: 'A_' matches Ab, Ac — NOT A or Abc
[Key Differences]
  - % matches any number of characters (0+); _ matches exactly 1 character
  - % is used for open-ended searches; _ is used for fixed-width pattern matching
  - They can be combined: '19__%' matches '19' + exactly 2 chars + anything after
'''


# ========================================================
# COMPARE: ️ LIKE vs NOT LIKE — Key Differences
# SESSION: Session 2 — WHERE Predicates, Set Ops & Functions
# ========================================================

'''
[LIKE]
  - Returns rows matching the specified pattern
  - Used for positive/inclusive pattern searches
  - Example: WHERE title LIKE '%1990%' — returns movies from 1990
[NOT LIKE]
  - Returns rows that do NOT match the specified pattern
  - Used for exclusion/negative pattern filtering
  - Example: WHERE title NOT LIKE '%19__%' — returns movies NOT from the 1900s
[Key Differences]
  - LIKE = include matching rows; NOT LIKE = exclude matching rows
  - Both use the same wildcard characters (% and _)
  - NOT LIKE is the logical inverse of LIKE for the same pattern
'''


# ### Session 3 — Aggregate Functions, GROUP BY & HAVING


# ========================================================
# COMPARE: ️ WHERE vs HAVING — Key Differences
# SESSION: Session 3 — Aggregate Functions, GROUP BY & HAVING
# ========================================================

'''
[WHERE Clause]
  - Filters individual rows before grouping occurs
  - Applied before GROUP BY
  - Cannot use aggregate functions (COUNT, SUM, AVG, etc.) in WHERE condition
  - Works on non-aggregated column values
  - Example: WHERE salary > 5000 — filters rows before any grouping
[HAVING Clause]
  - Filters groups after grouping occurs
  - Applied after GROUP BY
  - MUST use aggregate functions in its condition
  - Works on aggregated group results
  - Example: HAVING SUM(salary) > 35000 — filters groups after aggregation
[Key Differences]
  - Execution order: WHERE → GROUP BY → HAVING → SELECT → ORDER BY
  - WHERE acts on rows; HAVING acts on groups
  - WHERE cannot use aggregates; HAVING must use aggregates
  - WHERE comes before GROUP BY; HAVING comes after GROUP BY
  - HAVING without GROUP BY treats the whole table as one group
  NOTE: ⚠️ Most Common Exam Trap: Students write aggregate conditions in the WHERE clause. Example: WHERE SUM(salary) > 10000 is WRONG — always use HAVING for conditions on aggregate results.
'''


# ========================================================
# COMPARE: ️ COUNT vs SUM vs AVG vs MIN vs MAX — Quick Reference
# SESSION: Session 3 — Aggregate Functions, GROUP BY & HAVING
# ========================================================

'''
[COUNT]
  - Returns the number of rows (count)
  - COUNT(*) includes NULL rows; COUNT(col) excludes NULL rows
  - Use when: you want to know how many records exist
[SUM]
  - Returns the total sum of a numeric column
  - Ignores NULL values
  - Use when: you want the total (total salary, total sales, total quantity)
[AVG]
  - Returns the arithmetic mean (average) of a numeric column
  - Ignores NULL values — only non-NULL values are averaged
  - Use when: you want the central tendency (average salary, average price)
[MIN]
  - Returns the smallest value in a column
  - Works on numeric, string, and date columns
  - Use when: you want the lowest value (minimum salary, earliest date, cheapest price)
[MAX]
  - Returns the largest value in a column
  - Works on numeric, string, and date columns
  - Use when: you want the highest value (maximum salary, latest date, most expensive item)
[Key Differences]
  - COUNT counts rows; SUM/AVG/MIN/MAX operate on column values
  - COUNT(*) includes NULLs; all others (SUM, AVG, MIN, MAX, COUNT(col)) ignore NULLs
  - SUM and AVG require numeric columns; MIN and MAX work on any data type
  - All five return zero when no matching rows exist
'''


# ### Session 4 — JOINs


# ========================================================
# COMPARE: ️ All JOIN Types — Complete Comparison
# SESSION: Session 4 — JOINs
# ========================================================

'''
[1. INNER JOIN]
  - Returns: Only rows where the join condition is TRUE in BOTH tables
  - NULLs in output: No — unmatched rows from either side are excluded entirely
  - Keyword: INNER JOIN … ON
  - Best for: Fetching records that are definitively linked (customer + their accounts)
  - Bank example: Customers who have accounts — customers with no accounts are excluded
[2. LEFT JOIN (Left Outer Join)]
  - Returns: All rows from the left table + matching rows from the right table
  - NULLs in output: Yes — right-table columns are NULL where no match exists
  - Keyword: LEFT JOIN … ON (OUTER is optional)
  - Best for: Finding records in the left table that have NO counterpart in the right table
  - Bank example: Finding dormant accounts — accounts with no transactions at all
[3. RIGHT JOIN (Right Outer Join)]
  - Returns: Matching rows from the left table + all rows from the right table
  - NULLs in output: Yes — left-table columns are NULL where no match exists
  - Keyword: RIGHT JOIN … ON (OUTER is optional)
  - Best for: Finding records in the right table that have NO counterpart in the left table
  - Bank example: Finding orphaned transactions — transactions whose account has expired
[4. FULL OUTER JOIN]
  - Returns: All rows from both tables; NULLs fill in where no match exists on either side
  - NULLs in output: Yes — on both left and right sides where no match
  - Keyword (MySQL): LEFT JOIN UNION RIGHT JOIN — not natively supported in MySQL
  - Best for: Complete data quality audits — see all unmatched records from both tables simultaneously
  - Bank example: View all accounts without transactions AND all transactions without accounts in one result
[5. NATURAL JOIN]
  - Returns: Rows where ALL common column names match (implicit join)
  - NULLs in output: No — NULLs in joining key columns cause rows to be silently dropped
  - Keyword: NATURAL JOIN — no ON clause needed
  - Best for: Clean, curated data where column names are designed to be unique across tables
  - Risk: Dangerous if tables have duplicate column names — can produce too many or too few rows
[6. CROSS JOIN (Cartesian Product)]
  - Returns: Every combination of every row — M rows × N rows = M×N total rows
  - NULLs in output: No
  - Keyword: CROSS JOIN — no ON clause
  - Best for: Broadcasting all items from one table to all rows of another (mass notifications, pricing matrices)
  - Risk: Exponential row growth on large tables — use with extreme caution
[7. SELF JOIN]
  - Returns: Rows by joining a table to itself using two different aliases
  - NULLs in output: Depends on the join condition applied (ON with date comparison, etc.)
  - Keyword: JOIN … AS alias1 ON alias1.col = alias2.col
  - Best for: Hierarchical/comparative data within a single table (parent–child, current vs previous month)
  - Bank example: Compare current month transactions to previous month transactions from the same table
[8. Non-Equi Join]
  - Returns: Rows matched using non-equality operators: >, <, <>, BETWEEN, LIKE, NOT LIKE
  - NULLs in output: No — only qualifying rows are returned based on the condition
  - Keyword: JOIN … WHERE col > value
  - Best for: Range-based lookups, tier pricing, interest rate classification, anomaly detection
  - Bank example: Find accounts eligible for privileged interest rate (rate > 7%) excluding credit cards
'''


# ========================================================
# COMPARE: ️ INNER JOIN vs OUTER JOINs (LEFT / RIGHT / FULL) — Key Differences
# SESSION: Session 4 — JOINs
# ========================================================

'''
[INNER JOIN]
  - Only rows that satisfy the join condition in BOTH tables are returned
  - Unmatched rows from either table are completely excluded from the result
  - Result set is always a subset of both tables
  - No NULL values produced from unmatched rows
  - Use when: you only care about records that have valid links in both tables
[OUTER JOINs (LEFT / RIGHT / FULL)]
  - Returns ALL rows from the specified table(s) even if no match exists
  - Unmatched rows appear with NULL in the columns of the non-matching table
  - Result set can be larger than an INNER JOIN result
  - NULLs in output signal missing/unmatched relationships
  - Use when: you need to identify records that have NO counterpart in the other table
[Key Differences]
  - INNER JOIN excludes unmatched rows; OUTER JOINs include unmatched rows with NULL fill
  - INNER JOIN result is a strict intersection; OUTER JOIN result can include exclusive rows from one or both tables
  - OUTER JOINs produce NULLs in the result; INNER JOIN never produces NULL from unmatched rows
  - For the same data, INNER JOIN always returns fewer or equal rows than any OUTER JOIN
  NOTE: ⚠️ Common Exam Trap: Students say INNER JOIN is the same as NATURAL JOIN. They are NOT the same — INNER JOIN requires an explicit ON condition and handles NULLs; NATURAL JOIN is implicit and drops NULL-key rows silently.
'''


# ========================================================
# COMPARE: ️ LEFT JOIN vs RIGHT JOIN — Key Differences
# SESSION: Session 4 — JOINs
# ========================================================

'''
[LEFT JOIN]
  - The left table is the "dominant" table — all its rows always appear in the output
  - The right table is the "optional" table — NULLs appear where no match exists
  - Prioritises the table listed first in the FROM clause
  - Use when: you want all records from the primary/master table and optional data from a secondary table
  - Bank example: ACCOUNT LEFT JOIN TRANSACTION → all accounts shown, even those with no transactions (dormant accounts)
[RIGHT JOIN]
  - The right table is the "dominant" table — all its rows always appear in the output
  - The left table is the "optional" table — NULLs appear where no match exists
  - Prioritises the table listed after the JOIN keyword
  - Use when: you want all records from a secondary/event table and optional data from the primary table
  - Bank example: ACCOUNT RIGHT JOIN TRANSACTION → all transactions shown, even those whose account is expired (orphaned transactions)
[Key Differences]
  - LEFT JOIN = all left rows + matching right rows; RIGHT JOIN = matching left rows + all right rows
  - Which table produces NULLs: in LEFT JOIN it's the right table; in RIGHT JOIN it's the left table
  - Every RIGHT JOIN can be rewritten as a LEFT JOIN by swapping the table order — they are logically equivalent
  - Use LEFT JOIN to find records in A with no match in B; use RIGHT JOIN to find records in B with no match in A
  NOTE: 💡 Best Practice: Many developers always use LEFT JOIN (never RIGHT JOIN) and simply swap the table order when needed. This keeps queries more readable and consistent.
'''


# ========================================================
# COMPARE: ️ NATURAL JOIN vs EQUI-JOIN (INNER JOIN) — Key Differences
# SESSION: Session 4 — JOINs
# ========================================================

'''
[NATURAL JOIN]
  - Joining columns are identified implicitly — MySQL scans for all common column names automatically
  - No ON or WHERE clause needed for the join condition
  - The common column appears only once in the output
  - Drops rows silently when the joining key column contains NULL
  - Risky: if extra columns share a name accidentally, the join silently adds incorrect conditions
  - Use when: tables are designed with clean, purposeful column naming and you want brevity
[EQUI-JOIN (INNER JOIN with =)]
  - Joining columns are specified explicitly using ON or WHERE clause with = operator
  - Requires ON clause: ON table1.col = table2.col
  - Both copies of the joining column appear in the output (unless you use USING)
  - Handles NULL values better — can use IFNULL() to substitute defaults
  - Safer and more predictable — developer controls exactly which columns are matched
  - Use when: you need control, NULL handling, or when column names are not identical across tables
[Key Differences]
  - NATURAL JOIN is implicit (auto-detects columns); EQUI-JOIN is explicit (developer specifies columns)
  - NATURAL JOIN is more compact; EQUI-JOIN is more readable and safer
  - NATURAL JOIN silently drops NULL-key rows; EQUI-JOIN with IFNULL() can preserve them
  - EQUI-JOIN with USING clause is a middle ground — explicit column, but appears once like NATURAL JOIN
  NOTE: ⚠️ Common Mistake: Assuming NATURAL JOIN and INNER JOIN always produce the same result. They will differ if NULL values exist in the join key or if multiple columns share the same name accidentally across tables.
'''


# ========================================================
# COMPARE: ️ Equi-Join vs Non-Equi Join — Key Differences
# SESSION: Session 4 — JOINs
# ========================================================

'''
[Equi-Join]
  - Uses the equality operator (=) to match rows between tables
  - Joins tables based on exact column value matches
  - Most common type of join — used for PK–FK relationships
  - Produces predictable one-to-one or one-to-many matching
  - Bank example: ACCOUNT.Acct_Num = TRANSACTION.Acct_Num (each transaction linked to one account)
[Non-Equi Join]
  - Uses comparison operators other than = : >, <, <>, BETWEEN, LIKE, NOT LIKE
  - Joins tables based on range conditions, pattern matching, or inequality
  - Does not require a direct FK–PK relationship
  - Produces cross-dimensional matching — often creates a many-to-many-like result
  - Bank example: ACCOUNT JOIN INTEREST WHERE IR.RATE > 0.07 (assigns privileged rate to qualifying account types)
[Key Differences]
  - Equi-Join uses =; Non-Equi Join uses >, <, BETWEEN, LIKE, NOT LIKE, <>
  - Equi-Join is based on FK–PK relationships; Non-Equi Join is based on business range/pattern logic
  - Equi-Join result size is bounded by the FK cardinality; Non-Equi Join can produce many more rows
  - Non-Equi Join is ideal for range lookups (interest tiers, discount bands, scoring ranges)
'''


# ### Session 5 — Subqueries


# ========================================================
# COMPARE: ️ IN vs EXISTS — Key Differences
# SESSION: Session 5 — Subqueries
# ========================================================

'''
[IN Operator]
  - How it works: The subquery executes first and returns a complete list of values; the outer query then checks membership for every row
  - Execution: Subquery runs once — its full result list is evaluated against the outer query
  - NULL behaviour: If the subquery returns even one NULL, NOT IN returns no rows (NULL breaks the logic). IN itself ignores NULLs for positive matches
  - Return type: Multi-row — works with a list of values
  - Best for: When the subquery result set is small and you need exact value matching
  - Bank example: WHERE Acct_Num IN (SELECT Acct_Num FROM TRANSACTION WHERE Channel = 'ATM withdrawal')
[EXISTS Operator]
  - How it works: For each outer row, MySQL runs the inner query — if it returns at least one row, EXISTS = TRUE
  - Execution: Subquery runs once per outer row — it is always correlated
  - NULL behaviour: Safe with NULLs — EXISTS only checks for row presence, not value equality
  - Return type: Boolean (TRUE / FALSE) — the actual value selected inside doesn't matter
  - Best for: Checking whether a relationship exists rather than fetching a list; efficient when the subquery result would be very large
  - Bank example: WHERE EXISTS (SELECT 'yes' FROM TRANSACTION T WHERE T.Acct_Num = A.Acct_Num)
[Key Differences]
  - Execution count: IN subquery runs once; EXISTS subquery runs once per outer row
  - NULL safety: EXISTS is safe with NULLs; NOT IN is dangerous if subquery can return NULL
  - Return: IN returns a list; EXISTS returns only TRUE/FALSE
  - Columns: IN must match on the correct column; EXISTS uses SELECT 'yes' — the column doesn't matter
  - Performance: EXISTS is often faster for large subquery result sets (stops at first match); IN is faster for small result sets
  NOTE: ⚠️ Common Exam Trap: Using NOT IN when the subquery can return NULLs always produces 0 rows. Use NOT EXISTS instead in that scenario — it handles NULLs correctly.
'''


# ========================================================
# COMPARE: ️ ANY vs ALL — Key Differences
# SESSION: Session 5 — Subqueries
# ========================================================

'''
[ANY Operator]
  - Meaning: Condition is TRUE if it holds for at least one value in the subquery list
  - > ANY — effectively means "greater than the minimum " of the list
  - < ANY — effectively means "less than the maximum " of the list
  - Result size: Returns more rows — looser condition (easier to satisfy)
  - Bank example: Tran_Date > ANY (SELECT Event_dt FROM Message WHERE event = 'Holiday') → returns all transactions after the earliest holiday (2020-02-19)
  - Equivalent to: > MIN(subquery)
[ALL Operator]
  - Meaning: Condition must hold for every single value in the subquery list
  - > ALL — effectively means "greater than the maximum " of the list
  - < ALL — effectively means "less than the minimum " of the list
  - Result size: Returns fewer rows — stricter condition (harder to satisfy)
  - Bank example: Tran_Date > ALL (SELECT Event_dt FROM Message WHERE event = 'Holiday') → returns only transactions after the latest holiday (2020-04-24)
  - Equivalent to: > MAX(subquery)
[Key Differences]
  - ANY = "satisfies at least one" → equivalent to MIN for > or MAX for <
  - ALL = "satisfies every value" → equivalent to MAX for > or MIN for <
  - ANY is more permissive and returns more rows; ALL is more restrictive and returns fewer rows
  - With the same subquery: > ANY(list) always returns more rows than > ALL(list)
  - ANY is similar to IN for equality; ALL has no direct equivalent using IN
  NOTE: 💡 Memory trick: ANY = at least one match (minimum bar). ALL = must beat everyone (maximum bar). Think of ANY as passing if one judge approves; ALL as needing unanimous approval from all judges.
'''


# ========================================================
# COMPARE: ️ Correlated vs Non-Correlated Subquery — Key Differences
# SESSION: Session 5 — Subqueries
# ========================================================

'''
[Non-Correlated Subquery]
  - Independence: Executes completely independently of the outer query — no reference to outer table
  - Execution count: Runs once ; its result is computed and cached, then reused
  - Scope: Uses only columns from its own table(s)
  - Performance: More efficient — runs only once regardless of outer result size
  - Use case: Pre-computing a threshold, list, or aggregate that doesn't depend on each outer row
  - Example: WHERE Tran_Date = (SELECT MAX(Event_dt) FROM Message WHERE Event = 'Holiday') — the inner query doesn't reference any outer table column
[Correlated Subquery]
  - Independence: Depends on the outer query — references a column from the outer table inside its WHERE clause
  - Execution count: Runs once per row of the outer query — potentially thousands of times
  - Scope: Uses columns from both its own table AND the outer query table
  - Performance: Slower on large datasets — re-executed for every outer row
  - Use case: Comparing each outer row against a dynamically computed value specific to that row
  - Example: EXISTS (SELECT 'yes' FROM TRANSACTION T WHERE T.Acct_Num = A.Acct_Num) — A.Acct_Num is from the outer query
[Key Differences]
  - Non-correlated: independent, runs once, faster; Correlated: dependent, runs per outer row, slower
  - Correlated subqueries always reference at least one outer query column in their WHERE
  - EXISTS is always correlated; IN/scalar subqueries can be either correlated or non-correlated
  - Correlated HAVING subqueries compare per-group metrics against group-specific aggregates (e.g. per-product average)
  NOTE: ⚠️ Common Mistake: Treating all subqueries as independent. If the inner WHERE references an outer table column (e.g. P2.Product = P1.Product ), it IS correlated and re-executes for every group/row.
'''


# ========================================================
# COMPARE: ️ Nested IN Subquery vs Nested JOIN Subquery — Key Differences
# SESSION: Session 5 — Subqueries
# ========================================================

'''
[Nested IN Subquery]
  - Structure: Chain of IN operators — outer IN → middle IN → inner SELECT
  - Column access: The final result can only contain columns from the outermost table — inner subquery columns are not accessible
  - Execution: Innermost executes first → passes values up through IN → each level filters using that list
  - Use case: Filtering a single table using criteria derived from multiple other tables
  - Example: Get TRANSACTION rows for customers with positive-balance accounts — final result only has TRANSACTION columns
[Nested JOIN Subquery]
  - Structure: Subqueries nested inside each other's FROM clause, each aliased and JOINed
  - Column access: The final result can contain columns from every joined subquery level — all columns are accessible
  - Execution: Innermost subquery runs first → passed as derived table to middle JOIN → final result combines all levels
  - Use case: When you need information from multiple levels — e.g. customer info + account info + transaction info + event info all in one result
  - Example: Get Cust_id, Balance, Tran_Amount, AND Event all in one result using layered FROM-clause subqueries
[Key Differences]
  - Column visibility: IN nesting only exposes outermost columns; JOIN nesting exposes all levels' columns
  - Structure: IN nesting uses WHERE IN; JOIN nesting uses FROM + JOIN ON aliases
  - Use IN nesting when you only need to filter the outermost table
  - Use JOIN nesting when you need to display data from multiple nested tables
  - Both execute innermost first, but JOIN nesting gives far more flexibility in the SELECT list
'''


# ========================================================
# COMPARE: ️ Scalar (Single-Row) vs Multi-Row Subquery — Key Differences
# SESSION: Session 5 — Subqueries
# ========================================================

'''
[Scalar / Single-Row Subquery]
  - Returns: Exactly one value (one row, one column)
  - Operators used with: =, <, >, <=, >= (single-value comparison operators)
  - Error if violated: If the subquery returns more than one row with =, MySQL throws: "Subquery returns more than 1 row"
  - How to guarantee scalar: Wrap with MAX(), MIN(), COUNT(), or AVG() to collapse multiple rows to one
  - Locations used: WHERE clause, CASE statement inside SELECT, HAVING clause
  - Example: WHERE Tran_Date = (SELECT MAX(Event_dt) FROM Message WHERE Event = 'Holiday')
[Multi-Row Subquery]
  - Returns: Multiple rows (but typically one column)
  - Operators used with: IN, NOT IN, EXISTS, NOT EXISTS, ANY, ALL
  - Safe with multiple rows: These operators are designed to handle lists — no error for multiple rows
  - How it works: The outer query iterates over the returned list for set membership or range checking
  - Example: WHERE Acct_Num IN (SELECT Acct_Num FROM TRANSACTION WHERE Channel = 'ATM withdrawal')
[Key Differences]
  - Scalar returns one row; multi-row returns many rows
  - Scalar uses =, <, >; multi-row uses IN, EXISTS, ANY, ALL
  - Using = with a multi-row subquery causes a runtime error — always use MAX/MIN to ensure scalar
  - Multi-row subqueries require multi-row operators; mixing them causes errors
  NOTE: ⚠️ Most Common Runtime Error: "Subquery returns more than 1 row" — caused by using = with a subquery that can return multiple rows. Fix: wrap the subquery with MAX() or MIN(), or change = to IN.
'''


# ========================================================
# COMPARE: ️ Subquery Placement — WHERE vs FROM vs HAVING vs SELECT
# SESSION: Session 5 — Subqueries
# ========================================================

'''
[Subquery in WHERE Clause]
  - Purpose: Filters rows of the main table based on a derived condition
  - Operators: IN, NOT IN, EXISTS, NOT EXISTS, ANY, ALL, =, <, >
  - Execution: Can be independent (runs once) or correlated (runs per outer row)
  - Output: Controls which rows are included — does not add new columns
  - Example: WHERE Acct_Num IN (SELECT Acct_Num FROM TRANSACTION WHERE Channel = 'ATM withdrawal')
[Subquery in FROM Clause (Derived Table)]
  - Purpose: Creates a temporary virtual table that the outer query can JOIN or reference
  - Must be aliased: (SELECT ...) alias_name — alias is mandatory
  - Execution: Always independent — runs once before the outer query begins
  - Output: Adds a fully queryable dataset — outer query can SELECT columns from it
  - Example: FROM (SELECT Acct_Num, Tran_Amount FROM TRANSACTION WHERE Tran_amount > 23000) Sub_T
[Subquery in HAVING Clause]
  - Purpose: Filters grouped results using a dynamically computed aggregate threshold
  - Requires: GROUP BY must be present; HAVING condition must use an aggregate function
  - Execution: Can be non-correlated (runs once) or correlated (runs once per group)
  - Output: Filters which groups are shown — does not add new columns
  - Example: HAVING SUM(Estimated_profit) > (SELECT AVG(estimated_profit) FROM Profit_Loss)
[Subquery in SELECT Clause (Scalar)]
  - Purpose: Adds a derived column value to each row — like a computed column from another table
  - Constraint: Must return exactly one row and one column (scalar)
  - Execution: Runs once per output row — behaves like a correlated subquery
  - Output: Adds a new column to the result set, computed per row
  - Example: CASE WHEN Tran_Date = (SELECT MAX(Event_dt) FROM Message) THEN 'HOLIDAY' ELSE 'WORKING DAY' END AS Event
[Subquery in WITH Clause (CTE)]
  - Purpose: Defines a named, reusable subquery that can be referenced multiple times in the main query
  - Execution: Runs once; result is cached and reused wherever the CTE name appears
  - Output: Acts like a temporary named table — JOINable, filterable, referenceable
  - Example: WITH I_RATE AS (SELECT Acct_type, rate FROM Interest) SELECT ... FROM ACCOUNT JOIN I_RATE ON ...
[Key Differences]
  - WHERE — filters rows; FROM — creates a joinable dataset; HAVING — filters groups; SELECT — adds a derived column
  - FROM subquery MUST have an alias; others do not require one
  - SELECT and correlated HAVING subqueries run once per row/group — avoid on very large tables
  - WITH clause is the only one that can be reused multiple times in the same query
  - Execution order: WITH (if present) → FROM subqueries → WHERE subqueries → HAVING subqueries → SELECT subqueries
'''


# ========================================================
# COMPARE: ️ Subquery vs JOIN — Complete Comparison
# SESSION: Session 5 — Subqueries
# ========================================================

'''
[Subquery]
  - Structure: A SELECT nested inside another SELECT — two separate logical units
  - Readability: Easier to read when the logic is complex — each piece is isolated and debuggable independently
  - Column access: IN-based subqueries can only return columns from the outer query table
  - Filtering: Pre-filters data inside the subquery before exposing it to the outer query
  - Correlated variant: Can be correlated (re-executes per outer row) for dynamic per-row calculations
  - Best for: Complex filtering logic, existence checks (EXISTS), aggregation comparisons (HAVING), when only outer table columns are needed in output
  - Example: WHERE Tran_Date IN (SELECT Event_dt FROM Message WHERE Event_dt BETWEEN '2020-04-01' AND '2020-06-01')
[JOIN]
  - Structure: Two or more tables combined in a single FROM clause with ON or WHERE condition
  - Readability: More compact but can become complex with many tables — all logic in one query
  - Column access: All columns from all joined tables are accessible in SELECT
  - Filtering: Filtering happens through ON and WHERE — same query level
  - Performance: Generally faster for simple relationship traversal — optimiser handles JOINs very efficiently
  - Best for: When columns from multiple tables are needed in the output, simple relationship traversal
  - Example: FROM TRANSACTION, Message WHERE Tran_Date = Event_dt AND Event_dt BETWEEN '2020-04-01' AND '2020-06-01'
[Key Differences]
  - Column output: JOIN exposes all joined columns; IN-subquery only exposes outer table columns
  - Readability: Subqueries separate logic into independent units; JOINs keep everything in one flat query
  - Performance: JOINs are often faster for large datasets; correlated subqueries can be slow (run per row)
  - Calculations: Subqueries return a single derived value from inner logic; JOINs calculate in the SELECT clause
  - Pre-filtering: Subqueries in FROM can pre-filter before a join; JOINs filter through WHERE/ON on the full result
  - Use subqueries when: logic is complex, only outer columns needed, checking existence, comparing aggregates
  - Use JOINs when: columns from multiple tables needed in output, simple key relationships, performance is critical
  NOTE: 💡 Best Practice: Subqueries and JOINs can be combined. Use a subquery in FROM to pre-filter a large table, then JOIN the result with another table for maximum flexibility and performance.
'''


# ### Session 6 — Window Functions, Normalization & Constraints


# ========================================================
# COMPARE: ️ RANK() vs DENSE_RANK() vs PERCENT_RANK() — Key Differences
# SESSION: Session 6 — Window Functions, Normalization & Constraints
# ========================================================

'''
[RANK()]
  - Assigns the same rank to tied rows , then skips subsequent ranks equal to the number of ties
  - Tie at rank-1 (2 rows) → next row gets rank-3 (rank-2 is skipped)
  - Gaps appear in the rank sequence whenever there are ties
  - Best for: competition-style ranking where tied positions mean the same outcome and the next rank must account for all tied entries
  - Example: 2 rows tie at rank-1 → next distinct row gets rank-3 (not rank-2)
[DENSE_RANK()]
  - Assigns the same rank to tied rows , but never skips the next rank
  - Tie at rank-1 (2 rows) → next row gets rank-2 (no gap)
  - Rank sequence is always consecutive — 1, 2, 3, 4... regardless of ties
  - Best for: when you want to know how many distinct performance levels exist (e.g., top 3 salary bands, not top 3 individual salaries)
  - Example: 2 rows tie at rank-1 → next distinct row gets rank-2 (not rank-3)
[PERCENT_RANK()]
  - Assigns a fractional rank between 0 and 1 using the formula: (rank − 1) / (rows − 1)
  - First row in a partition always gets 0 ; last row always gets 1
  - Like RANK(), it also skips the percentage rank for tied values
  - Best for: percentile-based analysis — where does this row stand relative to the entire dataset?
  - Example with 4 rows (no ties): 0, 0.333, 0.667, 1.0
[Key Differences]
  - Ties: All three assign same rank to ties — but only DENSE_RANK avoids a gap in the sequence after
  - Output range: RANK/DENSE_RANK → integers; PERCENT_RANK → 0.0 to 1.0 fraction
  - First row: RANK and DENSE_RANK start at 1; PERCENT_RANK first row is always 0
  - Sequence gaps: RANK has gaps; DENSE_RANK never has gaps; PERCENT_RANK has fractional gaps like RANK
  - Use RANK for competition; DENSE_RANK for band/tier classification; PERCENT_RANK for percentile analysis
  NOTE: ⚠️ Most Common Exam Trap: Confusing RANK and DENSE_RANK on ties. Remember: RANK skips, DENSE_RANK never skips. Both assign the same rank to tied rows.
'''


# ========================================================
# COMPARE: ️ LAG() vs LEAD() — Key Differences
# SESSION: Session 6 — Window Functions, Normalization & Constraints
# ========================================================

'''
[LAG(col)]
  - Looks backward — returns the value of col from the previous row in the ordered result
  - The first row has no previous row → its LAG value is NULL
  - Best for: comparing current record against the prior record (e.g., today's balance vs yesterday's)
  - Example: LAG(Balance) OVER(ORDER BY Balance) — each row shows the balance of the row before it
[LEAD(col)]
  - Looks forward — returns the value of col from the next row in the ordered result
  - The last row has no next row → its LEAD value is NULL
  - Best for: lookahead analysis (e.g., what is the next day's transaction amount, next period's target)
  - Example: LEAD(Balance) OVER(ORDER BY Balance) — each row shows the balance of the row after it
[Key Differences]
  - Direction: LAG goes backward (previous row); LEAD goes forward (next row)
  - NULL position: LAG → NULL on first row; LEAD → NULL on last row
  - Both return NULL when there is no previous/next row — there is no error
  - Both can accept an offset: LAG(col, 2) returns 2 rows back; LEAD(col, 2) returns 2 rows forward
  - Both can accept a default: LAG(col, 1, 0) returns 0 instead of NULL when no prior row exists
'''


# ========================================================
# COMPARE: ️ FIRST_VALUE() vs LAST_VALUE() — Key Differences
# SESSION: Session 6 — Window Functions, Normalization & Constraints
# ========================================================

'''
[FIRST_VALUE(col)]
  - Returns the first value from the ordered window frame
  - Works correctly with just ORDER BY — no special frame clause needed
  - The first value in the ordered set is repeated across all rows in the partition
  - Best for: "show the minimum (first ordered) value alongside every row" without using MIN() and losing row detail
  - Example: FIRST_VALUE(Balance) OVER(ORDER BY Balance) → shows the lowest balance on every row
[LAST_VALUE(col)]
  - Returns the last value from the ordered window frame
  - Requires the RANGE BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING frame clause to work correctly
  - Without the frame clause, MySQL defaults the frame end to the current row — so LAST_VALUE returns the current row's value (not the true last)
  - Best for: "show the maximum (last ordered) value alongside every row"
  - Example: LAST_VALUE(Balance) OVER(ORDER BY Balance RANGE BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) → shows the highest balance on every row
[Key Differences]
  - FIRST_VALUE: works out of the box; LAST_VALUE: requires explicit RANGE frame clause
  - Both repeat a single reference value across all rows in the partition
  - FIRST_VALUE picks from the start of ordered window; LAST_VALUE picks from the end
  - Without RANGE frame, LAST_VALUE gives wrong results — this is the single most common mistake with window functions
  NOTE: ⚠️ Critical Exam Note: LAST_VALUE without RANGE BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING returns the current row's value — NOT the true last. Always include the frame clause with LAST_VALUE.
'''


# ========================================================
# COMPARE: ️ NTILE() vs CUME_DIST() — Key Differences
# SESSION: Session 6 — Window Functions, Normalization & Constraints
# ========================================================

'''
[NTILE(n)]
  - Divides rows into exactly n buckets (equal-sized groups)
  - Returns an integer bucket number from 1 to n
  - If rows are not evenly divisible by n, the last bucket gets the extra rows
  - No concept of cumulative distribution — purely a bucketing/segmentation tool
  - Best for: segmentation analysis (top quartile, bottom quartile, tier classification)
  - Example: NTILE(4) → splits rows into 4 equal quartile buckets
[CUME_DIST()]
  - Returns the cumulative distribution percentage — the fraction of rows at or below the current row
  - Returns a decimal between 0 and 1 (never 0 — first row is at least 1/n)
  - Tied rows receive the same cumulative distribution value
  - Best for: "what percentage of records have a value ≤ this one?" — frequency distribution analysis
  - Example: if 4 rows with distinct values → CUME_DIST values are 0.25, 0.50, 0.75, 1.0
[Key Differences]
  - Output type: NTILE → integer (bucket number); CUME_DIST → decimal (0 to 1)
  - Purpose: NTILE = segmentation into equal groups; CUME_DIST = cumulative frequency percentage
  - First row: NTILE starts at 1; CUME_DIST starts at 1/n (never 0, unlike PERCENT_RANK)
  - Tied rows: NTILE assigns same bucket; CUME_DIST assigns same cumulative value
  - PERCENT_RANK first row = 0 always; CUME_DIST first row = 1/n (always greater than 0)
'''


# ========================================================
# COMPARE: ️ Window Function (OVER) vs GROUP BY — Full Comparison
# SESSION: Session 6 — Window Functions, Normalization & Constraints
# ========================================================

'''
[GROUP BY]
  - Collapses multiple rows into a single summary row per group
  - Individual row details are lost — only the aggregate is visible
  - Result set has fewer rows than the input (one per group)
  - Cannot show both individual row detail AND the group aggregate in the same result
  - Use when: you need a summary report — total per department, average per month
  - Example: SELECT dept, SUM(salary) FROM employee GROUP BY dept — one row per department
[Window Function with OVER()]
  - Preserves all rows in the result — no row is collapsed
  - The aggregate result is computed per group and returned alongside each individual row
  - Result set has the same number of rows as the input
  - Can show both individual row detail AND the group aggregate in the same output row
  - Use when: you need row-level detail alongside a group summary — each employee's salary next to their department total
  - Reduces need for self-JOINs — window functions achieve the same row comparison internally
  - Example: SUM(Balance) OVER(PARTITION BY Cust_Id) — each account row shows the customer's total balance
[Key Differences]
  - Row count: GROUP BY reduces rows; OVER() preserves all rows
  - Row detail: GROUP BY loses it; OVER() keeps it
  - Performance: OVER() reduces self-JOINs — better CPU utilization for row-level comparisons
  - Both can use aggregate functions (SUM, AVG, MIN, MAX, COUNT) but produce different shapes of output
  - Can be combined: GROUP BY for summary reports, window functions for analytical drill-down
'''


# ========================================================
# COMPARE: ️ 1NF vs 2NF vs 3NF vs BCNF — Normalization Compared
# SESSION: Session 6 — Window Functions, Normalization & Constraints
# ========================================================

'''
[1NF — First Normal Form]
  - Prerequisite: None — it is the starting point
  - Violation: Columns contain multi-valued or composite attributes (more than one value per cell)
  - Fix: Decompose multi-valued columns — place each atomic value in a separate row
  - Rule enforced: Every cell must contain one and only one atomic (indivisible) value
  - Example fix: A branch column holding multiple IFSC codes → split into separate rows, one IFSC per row
[2NF — Second Normal Form]
  - Prerequisite: Must be in 1NF
  - Violation: Partial dependency — non-key attributes depend on only part of a composite primary key
  - Fix: Split the table; ensure each non-key column depends on the full primary key of its table
  - Rule enforced: All non-key attributes must be fully functionally dependent on the entire primary key
  - Example fix: Equipment_Lease (composite PK: Machine_ID + Invoice) — Machine Name only depends on Machine_ID → split into EQUIPMENT and LEASE tables
[3NF — Third Normal Form]
  - Prerequisite: Must be in 2NF
  - Violation: Transitive dependency — non-key attribute A depends on non-key attribute B, which depends on the primary key
  - Fix: Split the table to remove transitive chain — each table has non-key columns directly dependent only on their own PK
  - Rule enforced: No non-prime attribute may transitively depend on the super key through another non-prime attribute
  - Example fix: ACCOUNT_TRANSACTION — Tran_amount depends on Tran_id, which depends on Acct_Num → split into ACCOUNT and TRANSACTION
[BCNF — Boyce-Codd Normal Form]
  - Prerequisite: Must be in 3NF
  - Violation: Residual redundancy in 3NF — multiple sub-transactions per single transaction create repeated data
  - Fix: Split into Fact table (high-level unique records) and Dimension table (breakdown detail)
  - Rule enforced: Every determinant must be a candidate key
  - Example fix: A cheque for education splits into fees, tax, cess (dimensions) linked to a single Cheque Number (fact)
[Key Differences]
  - 1NF: fixes multi-value cells → atomic values per cell
  - 2NF: fixes partial dependency → full dependency on whole PK (only applies when composite PK exists)
  - 3NF: fixes transitive dependency → non-key columns depend only on PK, not on other non-key columns
  - BCNF: fixes remaining redundancy in 3NF → fact/dimension split eliminates repeated data from sub-transactions
  - Each normal form is a superset of the previous: BCNF ⊃ 3NF ⊃ 2NF ⊃ 1NF
  - 2NF is only relevant when a composite primary key exists — single-column PK tables skip directly to 3NF concerns
  NOTE: ⚠️ Most Common Exam Trap: Confusing 2NF and 3NF. Remember: 2NF removes partial dependency (on part of a composite key); 3NF removes transitive dependency (non-key depending on non-key).
'''


# ========================================================
# COMPARE: ️ NOT NULL vs CHECK vs UNIQUE vs PRIMARY KEY — Constraint Comparison
# SESSION: Session 6 — Window Functions, Normalization & Constraints
# ========================================================

'''
[NOT NULL]
  - Restricts: The column cannot contain a NULL (blank/unknown) value
  - Duplicates allowed: Yes — two rows can have the same non-null value
  - NULL allowed: No — the whole point is to prevent NULL
  - Creates index: No
  - Per table: Can be applied to multiple columns
  - Use when: A field must always have a value (e.g., mandatory name, mandatory transaction amount)
[CHECK]
  - Restricts: Column values must satisfy a specified boolean condition
  - Duplicates allowed: Yes — as long as each value passes the condition
  - NULL allowed: Yes (NULL passes CHECK unless explicitly restricted with NOT NULL)
  - Creates index: No
  - Use when: Values must be within a range or meet a business rule (e.g., transaction amount < 2,000,000)
[UNIQUE]
  - Restricts: All non-null values in the column must be distinct (no duplicates)
  - Duplicates allowed: No for non-null values; one NULL may be allowed per MySQL
  - NULL allowed: Yes — one NULL value is typically permitted
  - Creates index: Yes — creates a non-clustered index
  - Per table: Multiple UNIQUE constraints per table allowed
  - Use when: A column must have distinct values but is not the primary identifier (e.g., IFSC code, email address)
[PRIMARY KEY]
  - Restricts: Uniquely identifies each row — combines NOT NULL + UNIQUE in one constraint
  - Duplicates allowed: No
  - NULL allowed: No — PRIMARY KEY enforces NOT NULL automatically
  - Creates index: Yes — creates a clustered index (the physical row order in InnoDB)
  - Per table: Only ONE primary key per table
  - Use when: Defining the main row identifier — Cust_Id, Acct_Num, etc.
[Key Differences]
  - NULL: NOT NULL prevents NULL; CHECK allows NULL (unless combined with NOT NULL); UNIQUE allows one NULL; PRIMARY KEY never allows NULL
  - Uniqueness: NOT NULL and CHECK allow duplicates; UNIQUE and PRIMARY KEY enforce uniqueness
  - Index: PRIMARY KEY and UNIQUE create indexes; NOT NULL and CHECK do not
  - Count per table: PRIMARY KEY → only 1; UNIQUE → multiple allowed; NOT NULL and CHECK → multiple allowed
  - PRIMARY KEY = NOT NULL + UNIQUE + index; it is the superset of the other constraints on its column
  NOTE: ⚠️ Common Mistake: Thinking UNIQUE and PRIMARY KEY are the same. PRIMARY KEY also enforces NOT NULL and allows only one per table. UNIQUE can allow NULL and there can be multiple unique keys per table.
'''


# ========================================================
# COMPARE: ️ ON UPDATE CASCADE vs ON DELETE CASCADE — Key Differences
# SESSION: Session 6 — Window Functions, Normalization & Constraints
# ========================================================

'''
[ON UPDATE CASCADE]
  - Trigger: Fires when the Primary Key value in the parent table is updated
  - Effect on child: The corresponding Foreign Key values in the child table are automatically updated to the new PK value
  - Use when: Parent identifiers can change and child rows should follow automatically (e.g., renaming a machine ID)
  - Example: Update EQUIPMENT.MACHINE_ID from MAC01 → MAC09; all LEASE rows referencing MAC01 automatically update to MAC09
  - Without CASCADE: The UPDATE on the parent fails because the child FK still references the old value
[ON DELETE CASCADE]
  - Trigger: Fires when a parent row is deleted
  - Effect on child: All child rows that reference the deleted parent row are automatically deleted
  - Use when: Deleting a parent record should clean up all related child records (e.g., deleting a machine removes all its lease records)
  - Example: DELETE FROM EQUIPMENT WHERE MACHINE_ID = 'MAC02'; all LEASE rows for MAC02 are automatically deleted
  - Without CASCADE: The DELETE on the parent fails because child FK rows still reference it
[Key Differences]
  - Trigger: UPDATE CASCADE fires on PK update; DELETE CASCADE fires on parent row deletion
  - Child effect: UPDATE CASCADE updates FK values; DELETE CASCADE removes child rows entirely
  - Both must be set up through ALTER TABLE with DROP FOREIGN KEY + ADD FOREIGN KEY (cannot modify FK in-place)
  - Both can be combined on the same FK: ON UPDATE CASCADE ON DELETE CASCADE
  - If neither is set: any UPDATE or DELETE on a referenced parent row will fail with a constraint violation
  NOTE: 💡 To add CASCADE to an existing FK, you must first DROP the existing FK constraint by name, then re-add it with the CASCADE clauses. MySQL does not allow modifying an existing FK directly.
'''


# ========================================================
# COMPARE: ️ ACID Properties — Detailed Comparison
# SESSION: Session 6 — Window Functions, Normalization & Constraints
# ========================================================

'''
[A — Atomicity]
  - Guarantee: All operations in a transaction succeed together, or none of them do
  - Violation scenario: A transfers $1000 to B; A's account is debited but B's is not credited (system crash) → money disappears
  - Enforcement: ROLLBACK mechanism — if any statement fails, all prior statements in the transaction are undone
  - Keyword: All-or-Nothing
[C — Consistency]
  - Guarantee: Database moves from one valid state to another valid state — no data is leaked or corrupted
  - Violation scenario: A sends $2000 but B receives only $1000 — $1000 is unaccounted for (transaction appears to succeed but data is inconsistent)
  - Enforcement: Application-level rules and database constraints ensure correct transformation logic
  - Keyword: No data leakage
[I — Isolation]
  - Guarantee: Each transaction is isolated — concurrent transactions do not see each other's intermediate state
  - Violation scenario: Edward has $5000 and writes two cheques for $3000 and $2700 — if both process simultaneously, total withdrawal exceeds balance
  - Enforcement: Locking, serialization, and transaction isolation levels (READ COMMITTED, SERIALIZABLE, etc.)
  - Keyword: Transactions are serialized
[D — Durability]
  - Guarantee: Once committed, data is permanently written — it survives system crashes, power failures, or restarts
  - Violation scenario: Edward swipes $2000; payment goes through; network crashes; money is debited but order not placed — without durability, the transaction state is lost
  - Enforcement: Write-ahead logging (WAL), transaction logs, and checkpointing mechanisms
  - Keyword: Committed = permanent
[Key Differences]
  - Atomicity: about completeness — all or nothing at transaction level
  - Consistency: about correctness — no data imbalance even if the transaction succeeds
  - Isolation: about concurrency — transactions don't interfere with each other
  - Durability: about persistence — committed data is not lost even after failure
  - Atomicity prevents partial writes; Consistency prevents data imbalance; Isolation prevents race conditions; Durability prevents data loss
'''


# ### Session 7 — Transactions, Locking & Views


# ========================================================
# COMPARE: ️ COMMIT vs ROLLBACK — Key Differences
# SESSION: Session 7 — Transactions, Locking & Views
# ========================================================

'''
[COMMIT]
  - Purpose: Permanently saves all DML changes made in the current transaction to the physical database files
  - Visibility: After COMMIT, changes are visible to all other users and sessions
  - Reversibility: Cannot be undone — there is no undo after COMMIT
  - Effect on savepoints: All savepoints defined within the transaction are permanently cleared
  - Use when: All DML steps have succeeded and the final state should be made permanent
  - Example: Delete inactive accounts, verify the result, then COMMIT — the deletions are permanent
[ROLLBACK]
  - Purpose: Undoes all DML changes in the current (not yet committed) transaction; restores the original state of the data
  - Visibility: Reverts to the state that was visible before the transaction began
  - Reversibility: Only works before COMMIT — once committed, ROLLBACK has no effect on those changes
  - Partial rollback: ROLLBACK TO savepoint_name undoes only the portion after that savepoint
  - Use when: An error is detected mid-transaction, or the user wants to cancel all changes
  - Example: Delete a customer record, then realise it was wrong — ROLLBACK restores the customer
[Key Differences]
  - Permanence: COMMIT is irreversible; ROLLBACK is the safety net (but only before COMMIT)
  - Savepoints: COMMIT clears all savepoints; ROLLBACK TO a savepoint preserves earlier savepoints
  - State after: COMMIT → new state is final; ROLLBACK → original pre-transaction state restored
  - Timing: ROLLBACK must be used before COMMIT — you cannot rollback a committed transaction
  NOTE: ⚠️ Most Common Mistake: Attempting ROLLBACK after COMMIT. Once a transaction is committed, there is no database-level undo. Always verify with SELECT before COMMIT on destructive operations.
'''


# ========================================================
# COMPARE: ️ SAVEPOINT vs RELEASE SAVEPOINT — Key Differences
# SESSION: Session 7 — Transactions, Locking & Views
# ========================================================

'''
[SAVEPOINT label_name]
  - Purpose: Marks a named checkpoint within an active (uncommitted) transaction
  - Effect: Captures the current state of the transaction — all DML done so far is "safe" up to this point
  - ROLLBACK TO label_name: Undoes everything done after this savepoint; the savepoint itself and earlier savepoints still exist
  - Lifetime: Exists until the transaction COMMITs (all cleared) or is explicitly released
  - Use when: You want fine-grained error recovery — undo only part of a large transaction
[RELEASE SAVEPOINT label_name]
  - Purpose: Explicitly deletes a named savepoint from within the current transaction
  - Effect: The savepoint no longer exists — ROLLBACK TO that savepoint now throws an error
  - Transaction not affected: Releasing a savepoint does NOT rollback or commit — the transaction continues normally
  - Earlier savepoints: Savepoints created before the released one still work normally
  - Use when: A checkpoint is no longer needed — release it to reduce overhead
[Key Differences]
  - SAVEPOINT creates a checkpoint; RELEASE SAVEPOINT deletes it
  - ROLLBACK TO works on existing savepoints; fails on released ones
  - Releasing a savepoint does not affect transaction state; it only removes the marker
  - All savepoints disappear automatically when the transaction COMMITs — RELEASE is for explicit early removal
  - Chain rule: releasing sp2 does not affect sp1 (before it) or sp3 (after it)
'''


# ========================================================
# COMPARE: ️ REPEATABLE READ vs READ COMMITTED — Isolation Level Comparison
# SESSION: Session 7 — Transactions, Locking & Views
# ========================================================

'''
[REPEATABLE READ (Default in MySQL)]
  - Lock type: Table-level lock — the entire table is locked when a DML operation begins
  - Scope: ALL rows in the table are locked — including rows not being modified
  - Effect on other sessions: No UPDATE, INSERT, or DELETE from any other session on the table until the first transaction commits
  - Read consistency: Guarantees the same data will be seen on repeated reads within the transaction — no phantom rows
  - Trade-off: Highest integrity, lowest concurrency
  - Best for: Batch processing, financial transactions where data cannot change mid-transaction
[READ COMMITTED]
  - Lock type: Record-level lock — only the specific rows being modified are locked
  - Scope: Only the modified rows are locked; other rows remain accessible
  - Effect on other sessions: Other sessions can INSERT new records and update unlocked rows freely
  - Read consistency: Each read sees the latest committed data — a repeated SELECT may return different results if another session committed changes between reads
  - Trade-off: Higher concurrency, slightly lower integrity (non-repeatable reads possible)
  - Best for: High-throughput applications where record-level isolation is sufficient
[Key Differences]
  - Lock granularity: REPEATABLE READ → table; READ COMMITTED → record only
  - Concurrency: REPEATABLE READ blocks all other DML; READ COMMITTED allows most other DML
  - Phantom rows: REPEATABLE READ prevents them; READ COMMITTED allows them
  - INSERT from other sessions: blocked by REPEATABLE READ; allowed by READ COMMITTED
  NOTE: ⚠️ Key Exam Fact: REPEATABLE READ is MySQL's DEFAULT isolation level. READ COMMITTED must be set explicitly using SET TRANSACTION ISOLATION LEVEL READ COMMITTED .
'''


# ========================================================
# COMPARE: ️ READ WRITE vs READ ONLY — Transaction Modifiers
# SESSION: Session 7 — Transactions, Locking & Views
# ========================================================

'''
[READ WRITE (Default)]
  - Behaviour: All DML operations are permitted — SELECT, INSERT, UPDATE, DELETE
  - Default: This is the default mode — no explicit setup needed
  - Use when: The transaction needs to modify data (standard transactional use)
  - Syntax: SET TRANSACTION READ WRITE;
[READ ONLY]
  - Behaviour: Only SELECT is permitted — any UPDATE, INSERT, or DELETE throws an error: "Cannot execute statement in a READ ONLY transaction"
  - Use when: A reporting or audit session that must not modify data — enforces discipline
  - Session-level change: SET SESSION TRANSACTION READ ONLY takes effect from the next transaction , not the current one
  - Syntax: SET TRANSACTION READ ONLY;
[Key Differences]
  - READ WRITE allows all DML; READ ONLY allows only SELECT
  - READ WRITE is the default — no action needed; READ ONLY must be explicitly set
  - Session modifier timing: SET SESSION TRANSACTION change takes effect on the NEXT transaction — the current transaction continues in its current mode
  - READ ONLY is useful for creating a "view-only" session boundary without database-level permission changes
'''


# ========================================================
# COMPARE: ️ Shared Locks (S) vs Exclusive Locks (X) — Full Comparison
# SESSION: Session 7 — Transactions, Locking & Views
# ========================================================

'''
[Shared Lock (S) — Read Lock]
  - Acquired by: SELECT … LOCK IN SHARE MODE
  - Compatibility: Multiple sessions can hold S locks on the same record simultaneously — they do not conflict
  - Purpose: Signals intent to read and potentially update — prevents others from writing while reading
  - Escalation: When the same session follows with an UPDATE, the S lock escalates to an X lock on that record
  - Blocked by: An existing X lock on the same record — S lock request must wait in queue
[Exclusive Lock (X) — Write Lock]
  - Acquired by: INSERT, UPDATE, DELETE — automatically, no explicit syntax needed
  - Compatibility: Completely exclusive — only one X lock per record at a time
  - Purpose: Prevents any other transaction from reading or writing the record while it is being modified
  - Effect: When X lock is held, other sessions cannot acquire EITHER an S lock OR an X lock on the same record
  - Released by: COMMIT or ROLLBACK at the end of the transaction
[Key Differences]
  - S locks are compatible with each other; X locks are compatible with nothing
  - Trigger: S lock is explicit ( LOCK IN SHARE MODE ); X lock is implicit (any write DML)
  - Once X lock acquired: all other sessions — whether requesting S or X — must wait in queue
  - S lock promotes to X lock when followed by a write operation in the same session
  - Row-level vs table-level: depends on isolation level — READ COMMITTED locks records; REPEATABLE READ locks the table
  NOTE: ⚠️ Critical Rule: When Session-1 holds an Exclusive lock, Session-2 cannot get EITHER a Shared lock OR an Exclusive lock on the same record. Both types must wait until Session-1 releases the X lock.
'''


# ========================================================
# COMPARE: ️ All View Types — Complete Comparison
# SESSION: Session 7 — Transactions, Locking & Views
# ========================================================

'''
[1. Simple View]
  - Definition: Created from a SELECT on a single table
  - DML operations: Fully supported — INSERT, UPDATE, DELETE work like on the base table
  - Storage: Stored in data dictionary (not physical data)
  - Use when: You need a reusable query on one table with no complex joins
[2. Complex View]
  - Definition: Created from a SELECT with multiple tables (JOINs, subqueries, WITH clauses)
  - DML operations: Limited — DML through complex views is restricted in most cases
  - Storage: Stored in data dictionary
  - Use when: You want to encapsulate complex multi-table logic so the calling query stays clean
[3. Inline View]
  - Definition: A subquery placed directly in the FROM clause of a SELECT — not stored anywhere
  - DML operations: Not supported — inline views are read-only derived sets
  - Storage: NOT stored in data dictionary — defined fresh each time in the query
  - Use when: You need a one-time derived dataset for a single query; do not want to create a persistent view
[4. Materialized View (Oracle)]
  - Definition: Stores the physical result data of the query — also called a Snapshot
  - DML operations: Generally read-only; data is refreshed not updated directly
  - Storage: Stored physically on disk — uses actual storage space
  - Refresh: Must be explicitly refreshed — Complete (full reload) or Fast (delta only)
  - Use when: Complex queries run too slowly to execute on every call; centralized data marts for reporting
[Key Differences]
  - Storage: Simple/Complex/Group/Joined views store only the query definition; Materialized views store actual data; Inline views store nothing
  - DML: Simple view supports full DML; Complex view has restricted DML; Inline and Materialized are generally read-only
  - Performance: Materialized view is fastest for complex queries (pre-computed); other views re-execute the query each time
  - Maintenance: Inline view needs no maintenance; Materialized view needs refresh; others need updates when base schema changes
'''


# ========================================================
# COMPARE: ️ Horizontal vs Vertical View — Full Comparison
# SESSION: Session 7 — Transactions, Locking & Views
# ========================================================

'''
[Horizontal View]
  - Column selection: Uses wildcard SELECT * — all columns are included dynamically
  - Flexibility: Adding or renaming columns in the base table does NOT require updating the view
  - Execution plan: Recalculated on each access — columns are not cached because they are dynamic
  - Column hiding: Not possible — all columns are always exposed
  - Best for: Quick views where full column access is needed and the schema may change frequently
[Vertical View]
  - Column selection: Explicitly named columns — only listed columns are included
  - Flexibility: If base table columns are renamed, the view MUST be updated to match — otherwise an error occurs
  - Execution plan: Can be cached — known columns allow the planner to optimize access
  - Column hiding: Fully supported — omit sensitive columns (SSN, salary, phone) from the view
  - Best for: Security-sensitive views where certain columns must be hidden, or performance-critical views with stable schemas
[Key Differences]
  - Column definition: Horizontal = dynamic (*); Vertical = static (named columns)
  - Schema maintenance: Horizontal = zero; Vertical = required when column names change
  - Column security: Horizontal = cannot hide columns; Vertical = can hide sensitive columns
  - Execution plan: Horizontal recalculates; Vertical can cache — Vertical is more performant on stable schemas
'''


# ========================================================
# COMPARE: ️ LOCAL vs CASCADED CHECK OPTION — Full Comparison
# SESSION: Session 7 — Transactions, Locking & Views
# ========================================================

'''
[WITH LOCAL CHECK OPTION]
  - Scope: Enforces only this view's own WHERE condition
  - Parent views: Conditions of any parent/base views in the chain are ignored
  - Bypass risk: Data that violates the parent view's condition can be inserted through LOCAL — it is inserted into the base table but is invisible through the parent view
  - Use when: You only care about enforcing this view's local rule, not the full chain
  - Example: local_view (balance ≤ 90000) on normal_view (balance ≥ 80000) with LOCAL — INSERT of balance=65000 SUCCEEDS (only ≤ 90000 checked; ≥ 80000 bypassed)
[WITH CASCADED CHECK OPTION]
  - Scope: Enforces all conditions in the entire view chain — current view AND all parent views
  - Parent views: Conditions of ALL parent/base views are validated before INSERT succeeds
  - Bypass risk: None — every layer's condition must be satisfied
  - Use when: Strict data integrity is required across the entire view hierarchy
  - Example: cascaded_view (balance ≤ 90000) on normal_view (balance ≥ 80000) with CASCADED — only records with 80000 ≤ balance ≤ 90000 are accepted
[Key Differences]
  - LOCAL: enforces only this view; CASCADED: enforces all views in the hierarchy
  - LOCAL can bypass parent view conditions; CASCADED cannot
  - Data inserted through LOCAL may be invisible through parent views (inconsistent state)
  - CASCADED is the safer default — use LOCAL only when you have a deliberate reason to bypass parent conditions
  NOTE: ⚠️ Using LOCAL CHECK OPTION can result in data that is inserted into the base table but NOT visible through the parent view — a subtle data integrity issue. Prefer CASCADED for most use cases.
'''


# ========================================================
# COMPARE: ️ Simple vs Complex vs Inline vs Materialized View — Quick Lookup
# SESSION: Session 7 — Transactions, Locking & Views
# ========================================================

'''
[Simple View]
  - Based on: single table, no aggregates, no JOINs
  - DML: fully supported
  - Stored: yes (data dictionary)
  - Data stored physically: no
[Complex View]
  - Based on: multiple tables with JOINs, subqueries, or WITH clause
  - DML: restricted — generally read-only through the view
  - Stored: yes (data dictionary)
  - Data stored physically: no
[Inline View]
  - Based on: any SELECT query placed in the FROM clause of a parent query
  - DML: not supported — derived table, read-only
  - Stored: no — exists only for the duration of the parent query
  - Data stored physically: no
[Materialized View (Oracle)]
  - Based on: any complex SELECT — results are pre-computed and stored
  - DML: read-only; refresh required to update content
  - Stored: yes (data dictionary + physical storage)
  - Data stored physically: yes — actual data written to disk; requires additional storage
  - Performance: fastest for complex queries (pre-computed); Complete Refresh = full reload; Fast Refresh = delta only
[Key Differences Summary]
  - Persistence: Inline view is temporary (query-scoped); all others are persistent (stored in dictionary)
  - Physical data: only Materialized View stores actual data on disk
  - DML support: Simple View has fullest DML support; Materialized View has least
  - Performance: Materialized View (pre-computed) > Simple/Complex View (re-executed) > Inline View (always re-computed as part of parent query)
  - Maintenance: Materialized View needs explicit refresh; others are always in sync with the base table (Inline is always fresh)
'''
```
